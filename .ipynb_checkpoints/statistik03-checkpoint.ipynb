```python
#16
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import pylab as pylab
import statistics
data = {"Klasse_unten" : [50, 150, 200, 250], "Klasse_oben" : [150, 200, 250, 300], "abs_Haufigkeit" : [40, 10, 20, 50] }
df = pd.DataFrame(data)

ges_hauf = df["abs_Haufigkeit"].sum()
df["Kumulierte_hauf_abs"] = df["abs_Haufigkeit"].cumsum()
df["Rel_hauf"] = df["abs_Haufigkeit"] / ges_hauf

df["Klassenbreite"] = df["Klasse_oben"] - df["Klasse_unten"]
df["Klassenmitte"] = df["Klasse_unten"] + df["Klassenbreite"] / 2

df["Hauf_x_Mitte"] = df["abs_Haufigkeit"] * df["Klassenmitte"]

Mittel_ges = df["Hauf_x_Mitte"].sum() / ges_hauf
print("arithmetisches_Mittel ", round(Mittel_ges, 2))


#Medianklasse Kq = 2 Rel abs_hauf Summe durch 2 = 60 -> Kumulierte_hauf_abs ablesen
#Klasse_unten + ((Breite * (ges_hauf / 2 - KumHauf von Klasse 0 bis Kq -1 )) / abs_Haufigkeit Kq)

median = 200 + (( 50 * (ges_hauf/2 - 50)) / 20 )
print("Median ", round(median,2))

#df["(Mitte-Mittel_ges)2*hauf"] = ((df["Klassenmitte"] - Mittel_ges)**2) * df["abs_Haufigkeit"]
df["abs_hauf_x_(Mitte)^2"] = df["abs_Haufigkeit"] * (df["Klassenmitte"]**2)

varianz = df["abs_hauf_x_(Mitte)^2"].sum() / ges_hauf
print("Varianz ", round(varianz,2))

sd = math.sqrt(varianz)
print("SD ", round(sd,2))

#unter 20% Perzentil - rel_kum_hauf 1.Klasse die über 0,2 ligt (Einfallsklasse 0) - Perzentil_0,20 = Einfallsklasse_unten + (Einfallsklasse_breite / Relhauf_Einfallskl) * (Perzentil - Kum_rel_hauf von Klasse vor Einfallsklasse)
perzentil_20 = 50 + ( 100 / 0.333333 ) * (0.2 - 0)
print("20 % Perzentil ", round(perzentil_20, 2))

df.loc["Total"] = df.sum()
display(df)
```
