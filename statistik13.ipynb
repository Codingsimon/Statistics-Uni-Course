Gepaarte („verbundene“) Daten   
Gepaarter t-Test   

H0 : μd = 0  
H1 : μd ≠ 0  
- H0: µvor Training = µnach Training ➔ µvor Training − nach Training = 0
- H1: µvor Training ≠ µnach Training ➔ µvor Training − nach Training ≠ 0

### T(d) = d/sd * sqrt(n)   
mit d = (Σ x1 - x2)/n Summe aller Differenzen durch Anzahl -> Mittelwert    
   
((x1 - mean)**2 + (x2 - mean)**2 + (x3 - mean)**2) / (Anz_Messwerte - 1)  empirische Varianz    
sd = math.sqrt(empirische Varianz)    

| T(d ) | > tn−1,1−α/2 → H0 wird abgelehnt   
| T(d ) | ≤ tn−1,1−α/2 → H0 kann nicht abgelehnt werden

```python
# B1 Gepaarter t-Test verbundene Daten, Vorher nacher vergleich
import math
import pandas as pd
# a Hypothesen:
print("a H0: µLMP = µUS -> µLMP - µUS = 0")
print("  H1: µLMP != µUS -> µLMP - µUS != 0")

# b Teststatistik
data = {"Patientin" : [1, 2, 3, 4, 5, 6, 7, 8], "LMP" : [275, 292, 281, 284, 285, 283, 290, 294], "US" : [273, 285, 270, 272, 278, 276, 291, 290]}
df = pd.DataFrame(data)
df["Differenz"] = df["LMP"] - df["US"]

# mittelwert = DiffSumm / n
d_mittelwert = 49 / 8

# ((x1 - mean)**2 + (x2 - mean)**2 + (x3 - mean)**2) / (Anz_Messwerte - 1)  empirische Varianz
df["(x_-_mittel)^2"] = (df["Differenz"] - d_mittelwert)**2
var = 132.875000 # Summe von (x_-_mittel)^2
sd = math.sqrt(var)

# T(d) = d/sd * sqrt(n) 
Td = d_mittelwert/sd * math.sqrt(8) 
print("b Teststatistik: ", Td)

# c kritischer Wert
alpha = 0.05 #%
#tn−1, 1−α/2 = t   tn = Grundges
#t8-1, 1-0.05/2 = t7, 0.975
#nachsehen bei F(z) = 0.975 und 7 Freiheitsgraden => 2.37
print("c kritischer Wert: ", 2.37)

# d Testentscheidung
# | T(d ) | ≤ tn−1,1−α/2
print("d Testentscheidung | 1.5 | ≤ 2.37 → H0 kann nicht abgelehnt werden")

df.loc["Total"] = df.sum()
display(df)
```

```python
# B2 Gepaarter t-Test verbundene Daten, Vorher nacher vergleich
# a Hypothesen:
print("a H0: µVor = µNach -> µVor - µNach = 0")
print("  H1: µVor != µNach -> µVor - µNach != 0")

# b Teststatistik
data = {"Person" : [1, 2, 3, 4, 5, 6], "Vor" : [54.5, 60.4, 85.6, 78.2, 120.6, 120], "Nach" : [55.5, 69.6, 86.7, 81.6, 116.5, 121]}
df = pd.DataFrame(data)
df["Differenz"] = df["Vor"] - df["Nach"]

#mittelwert = DiffSum / n
d_mittelwert = -11.6 / 6

# ((x1 - mean)**2 + (x2 - mean)**2 + (x3 - mean)**2) / (Anz_Messwerte - 1)  empirische Varianz
df["(x_-_mittel)^2"] = (df["Differenz"] - d_mittelwert)**2
var = 93.793333 # Summe von (x_-_mittel)^2
sd = math.sqrt(var)

# T(d) = d/sd * sqrt(n) 
Td = d_mittelwert/sd * math.sqrt(8) 
print("b Teststatistik: ", Td)

# c kritischer Wert
alpha = 0.05 #%
#tn−1, 1−α/2 = t   tn = Grundges
#t6-1, 1-0.05/2 = t5, 0.975
#nachsehen bei F(z) = 0.975 und 5 Freiheitsgraden => 2.57
print("c kritischer Wert: ", 2.57)

# d Testentscheidung
# | T(d ) | ≤ tn−1,1−α/2
print("d | -0.5646 | ≤ 2.57 → H0 kann nicht abgelehnt werden")

df.loc["Total"] = df.sum()
display(df)
```

F-Test   
H0 : σX2 = σY2   
H1 : σX2 ≠ σY2  

F(x, y) = sdX²/sdY²  -> sdX immer größerer Wert als sdY

F(x, y) ≤ Fnx−1,n y−1,1−α → H0 kann nicht abgelehnt werden; Es liegt ein Hinweis auf Varianzhomogenität vor; unabhängige Zweistichproben-t-Test verwenden    
F(x, y) > Fnx−1,n y−1,1−α → H0 wird abgelehnt    

```python
# B3 Test auf Homoskedastizität F-Test
# a Hypothese
print("H0: µNeu = µBisher")
print("H1: µNeu ≠ µBisher")

# b Testalgorithmus
# F(x, y) = sdX²/sdY²   sqrt(var) = sd
Fxy = 49**2/9**2
print("b Teststatistik: ", Fxy)

# c kritischer Wert
nx = 63
ny = 70
alpha = 0.1 #%
# F nx−1, ny−1, 1-α
# F 63-1, 70-1, 1-0.1 = F 62, 69, 0.9
# F Verteilung nachsehen bei m = 62 Freiheitsgraden und n = 69 => 1.53
print("c kritischer Wert: ", 1.53)

# F(x, y) > Fnx−1, ny−1, 1−α → H0 wird abgelehnt
print("d 29.64 > 1.53 → H0 wird abgelehnt; Es liegt ein Hinweis auf Varianzhomogenität vor.")
```
