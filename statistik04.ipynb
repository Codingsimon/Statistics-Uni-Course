### Kovarianz   
Maßzahl für den Zusammenhang, gibt nur Richtung an nicht Stärke   
$Cov(x,y) = 1/N * ∑ (xi - x̄) * (yi − ȳ)$    
Cov(x,y) > 0 ... positiver Zusammenhang   
Cov(x,y) < 0 ... negativer Zusammenhang   
Cov(x,y) = 0 ... kein monotoner Zusammenhang   

### Stichproben-Korrelationskoeffizient

### Bravais-Pearson Korrelationskoeffizient r_xy   
metrisches Datenniveau und empfindlich gegen Ausreißer   
Kennziffer zur Beschreibung der Stärke und Art des Zusammenhangs; Wertebereich von -1 bis +1    
Um so näher der Betrag des Korrelationskoeffizienten bei 1 liegt, desto stärker ist der Zusammenhang.   
Korrelationskoeffizienten von 0 kein linearer Zusammenhang.   
$r_xy = {{Cov(x,y)}\over{\sqrt{Var(x) * Var(y)}}}$  
Berechnung mit dieser Formel:
<img src="Bravais-PearsonKorrelationskoeffizient.png" alt="Drawing" style="width: 600px;"/>     
rxy > 0 ... positiver linearer Zusammenhang   
rxy < 0 ... negativer linearer Zusammenhang   
rxy = 0 ... kein linearer Zusammenhang („unkorreliert“)   

### Regressionsgerade
$ b = Cov(x,y)/Var(x) = sd_xy/(sd_x)²$   
1/N kann weg gelassen werden da sich dies bei Cov und Var wegkürzen würde.   
$ a = ȳ − b ⋅ x̄ $   
$ yî = a + b ⋅ xi $   

```python
# 21 a Bravais-Pearson-Korrelationskoeffizienten
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = {"Person" : [1, 2, 3, 4, 5, 6], "Stunden_ohne_Schlaf" : [21, 20, 15, 26, 19, 12], "Programmierfehler" : [30, 32, 10, 40, 26, 23]}
df = pd.DataFrame(data)

x_mean = df["Stunden_ohne_Schlaf"].sum()/6
y_mean = df["Programmierfehler"].sum()/6
# Kovarianz (xi - x̄) * (yi − ȳ)
df["(x-x_mean)_*_(y-y_mean)"] = (df["Stunden_ohne_Schlaf"] - x_mean) * (df["Programmierfehler"] - y_mean)
df["(x-x_mean)^2"] = (df["Stunden_ohne_Schlaf"] - x_mean)**2
df["(y-y_mean)^2"] = (df["Programmierfehler"] - y_mean)**2

# Bravais_Pearson = ∑ (x-x_mean)_*_(y-y_mean) / sqrt(∑(x-x_mean)^2 * ∑(y-y_mean)^2)
Bravais_Pearson = 197.833333 / math.sqrt(118.833333 * 508.833333) # 0.80453
print("a Bravais-Pearson-Korrelationskoeffizienten rxy = ", Bravais_Pearson)

# b Regressionsgerade yî = a + b ⋅ xi
# b = Cov(x,y)/Var(x) => (x-x_mean)_*_(y-y_mean) / (x-x_mean)^2
b = 197.833333/118.833333  # = 1.66479663 

# a = ȳ − b ⋅ x̄
a = y_mean - b * x_mean # = -4.5203366

# yî = a + b ⋅ xi
print("b Regressionsgerade: y = -4.5203 + 1.6648 * x")

# c 13 ohne Schlaf 
x = 13
y = -4.5203 + 1.6648 * x
print("c es können nach 13 ohne Schlaft ", y ,"Fehler erwartet werden")

# plot Regressionsgerade
x = np.linspace(10,27,100)
y = -4.5203 + 1.6648 * x
df.plot.scatter(x='Stunden_ohne_Schlaf', y='Programmierfehler')
plt.plot(x, y, '-r', label='Regressionsgerade')
df.loc["Total"] = df.sum()
display(df)
```

### Gütemaßzahl für ein Regressionsmodell
### R² = (Bravais-Pearson Korrelationskoeffizient)²

Bestimmtheitsmaß R² = SQE/SQT   
Je größer das R2 ist, desto höher ist die Güte des Regressionsmodells.   
Wertebereich von 0 bis 1   
ab 0.60 brauchbares Modell   

Erklärte Streuung („Sum of Squares Explained“):   
SQE = ∑(^y - ȳ)² ; ^y = Schätzwert nach Regressionsfunktion    
Gesamtstreuung („Sum of Squares Total“):    
SQT = ∑(y - ȳ)²    

```python
# 22 a
# a Quatifizierung des Zusammenhangs: Bravais Pearson Korrelationskoeffizient
data = {"Blutdruck" : [146, 142, 149, 141, 132, 168, 170, 130], "Cholesterin" : [179, 202, 210, 189, 175, 203, 201, 165]}
df = pd.DataFrame(data)

x_mean = df["Blutdruck"].sum()/8
y_mean = df["Cholesterin"].sum()/8

# Kovarianz (xi - x̄) * (yi − ȳ)
df["(x-x_mean)_*_(y-y_mean)"] = (df["Blutdruck"] - x_mean) * (df["Cholesterin"] - y_mean)
df["(x-x_mean)^2"] = (df["Blutdruck"] - x_mean)**2
df["(y-y_mean)^2"] = (df["Cholesterin"] - y_mean)**2

# Bravais_Pearson = ∑ (x-x_mean)_*_(y-y_mean) / sqrt(∑(x-x_mean)^2 * ∑(y-y_mean)^2)
Bravais_Pearson  = 1172 / math.sqrt(1549.5 * 1804)
print("b Bravais-Pearson-Korrelationskoeffizienten rxy = ", Bravais_Pearson)

# b Cholesterinwert für Blutdruck von 160
# Regressionsgerade yî = a + b ⋅ xi
# b = Cov(x,y)/Var(x)
b = 1172 / 1549.5 # = 0.75637302355

# a = ȳ − b ⋅ x̄
a = y_mean - b * x_mean # = 79.124072

x = 160
y = a + b * x
print("b Colesterinwert ", y, " bei 160 Blutdruck")

# c Kennziffer für Güte dieses linearen Regressionsmodells = Bravais-Pearson-Korrelationskoeffizienten²
print("c Bestimmtheitsmaß R² = ", Bravais_Pearson**2)
print("Modell nicht brauchbar da Güte unter 0.6 ist")

df.loc["Total"] = df.sum()
display(df)
```

```python
# 24 a Korrelationskoeffizienten Bravais-Pearson
import math
data = {"Alter" : [10, 11, 15, 17, 16, 16, 6], "Zeit" : [180, 183, 119, 153, 43, 46, 250]}
df = pd.DataFrame(data)

mean_alter = df["Alter"].sum()/7 # 13
mean_zeit = df["Zeit"].sum()/7 # 139.14

# Kovarianz (xi - x̄) * (yi − ȳ)
df["(x-x_mean)_*_(y-y_mean)"] = (df["Alter"] - mean_alter) * (df["Zeit"] - mean_zeit)
df["(x-x_mean)^2"] = (df["Alter"] - mean_alter)**2
df["(y-y_mean)^2"] = (df["Zeit"] - mean_zeit)**2

# Bravais_Pearson = ∑ (x-x_mean)_*_(y-y_mean) / sqrt(∑(x-x_mean)^2 * ∑(y-y_mean)^2)
Bravais_Pearson = df["(x-x_mean)_*_(y-y_mean)"].sum() / math.sqrt(df["(x-x_mean)^2"].sum() * df["(y-y_mean)^2"].sum())
print("a Bravais-Pearson-Korrelationskoeffizienten rxy = ", Bravais_Pearson)
print("b rxy = ", Bravais_Pearson, " Korrelationskoeffizent nahe -1 daher, starker negativer linearer Zusammenhang")

# c Regressionsmodell mit Regressionsgerade
# b = Cov(x,y)/Var(x)  => (x-x_mean)_*_(y-y_mean) / (x-x_mean)^2    # var = sqrt(sd)
b = df["(x-x_mean)_*_(y-y_mean)"].sum()/df["(x-x_mean)^2"].sum() # = -15.39
# a = ȳ − b ⋅ x̄
a = mean_zeit - b * mean_alter # = 339.2128
# Regressionsgerade yî = a + b ⋅ xi
print("c Regressionsgerade y = 339.2128 + -15.39 * x")

# d  
x = 62
y = 339.2128 + -15.39 * x
print("d Erwarung bei 62 Jahren ", y)

x = np.linspace(5,18,100)
y = 339.2128 + -15.39 * x
df.plot.scatter(x='Alter', y='Zeit')
plt.plot(x, y, '-r', label='Regressionsgerade')
df.loc["Total"] = df.sum()
display(df)
```

### Rangkorrelationskoeffizient nach Spearman
Formel rxy,Spearman = Bravais-Pearson Formel mit Rängen   
misst nicht den linearen Zusammenhang sondern „generelle“ „Gleichläufigkeit“ bzw. „Gegenläufigkeit“ (Monotonieverhalten).   
nicht eindeutigen linearen Zusammenhangs    
- Gleichläufigkeit (monoton steigend): große x-Werte mit großen y-Werten   
- Gegenläufigkeit (monoton fallend): große x-Werte mit kleinen y-Werten   
+1 bzw. -1 liegt eine „perfekte“ Gleichläufigkeit    
0 besteht ein Zusammenhang der nicht ausschließlich gleichläufig bzw. gegenläufig    
Bei gleichem Rang Rangsumme bilden und Mittelwert für beide vergeben   

```python
# 26
data = {"Kaffeehaus" : [1, 2, 3, 4, 5, 6, 7], "A" : [6, 5, 8, 9, 7, 7, 9], "R(A)" : [ 2, 1, 5, 6.5, 3.5, 3.5, 6.5], "B" : [7, 4, 8, 10, 8, 6, 1], "R(B)" : [ 4, 2, 5.5, 7, 5.5, 3, 1]}
df = pd.DataFrame(data)

mean_RA = df["R(A)"].sum()/7
mean_RB = df["R(B)"].sum()/7

# Kovarianz (xi - x̄) * (yi − ȳ)
df["(x-x_mean)_*_(y-y_mean)"] = (df["R(A)"] - mean_RA) * (df["R(B)"] - mean_RB)
df["(x-x_mean)^2"] = (df["R(A)"] - mean_RA)**2
df["(y-y_mean)^2"] = (df["R(B)"] - mean_RB)**2

# Spearman = ∑ (x-x_mean)_*_(y-y_mean) / sqrt(∑(x-x_mean)^2 * ∑(y-y_mean)^2)
Spearman = 7.25 / math.sqrt(27 * 27.5)
print("a Rangkorrelationskoeffizient nach Spearman rxy,Spearman= ",Spearman)
print("b schwache gleichläufigkeit")

df.plot.scatter(x='R(A)', y='R(B)')
df.loc["Total"] = df.sum()
display(df)
```
